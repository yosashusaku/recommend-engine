```python
from recbole.quick_start import run_recbole

dataset='ml-100k'
model_list = [
  # General recommendation　27
  'Pop', 'ItemKNN', 'BPR', 'NeuMF', 'ConvNCF',
  'DMF', 'FISM', 'NAIS', 'SpectralCF', 'GCMC',
  'NGCF', 'LightGCN', 'DGCF', 'MultiVAE', 'MultiDAE',
  'CDAE', 'MacridVAE', 'LINE', 'EASE', 'RaCT',
  'RecVAE', 'NNCF', 'ENMF', 'SLIMElastic', 'SGL',
  'ADMMSLIM', 'NCEPLRec',
  # Context-aware recommendation 18
#  'LR', 'FM', 'NFM', 'DeepFM', 'xDeepFM',
#  'AFM', 'FFM', 'FwFM', 'FNN', 'PNN',
#  'DSSM', 'WideDeep', 'DIN', 'DCN', 'AutoInt',
#  'xgboost', 'lightgbm', 'DIEN',
  # Sequential recommendation 24
  'FPMC', 'GRU4Rec', 'NARM', 'STAMP', 'Caser',
  'NextItNet', 'TransRec', 'SASRec', 'BERT4Rec', 'SRGNN',
  'GCSAN', 'GRU4RecF', 'SASRecF', 'FDSA', 'GRU4RecKG',
  'KSR', 'FOSSIL', 'RepeatNet', 'SHAN', 'NPE',
  'HRM', 'HGN', 'LightSANs', 'SINE',
  # Knowledge-based recommendation 8
  'CKE', 'CFKG', 'KTUP', 'KGAT', 'RippleNet',
  'MKR', 'KGCN', 'KGNNLS'
]
model_neg_list = [
  'MultiVAE', 'MultiDAE', 'CDAE', 'MacridVAE', 'RecVAE',
  'ENMF', 'GRU4Rec', 'NARM', 'STAMP', 'Caser',
  'NextItNet', 'SASRec', 'BERT4Rec', 'SRGNN', 'GCSAN',
  'GRU4RecF', 'SASRecF', 'FDSA', 'S3Rec', 'GRU4RecKG',
  'KSR', 'FOSSIL', 'RepeatNet', 'SHAN', 'NPE',
  'HRM', 'LightSANs', 'SINE'
]
model_ract_list = ['RaCT']
model_s3rec_list = ['S3Rec']
```

```python
import time
import pandas as pd

recbole_result = pd.DataFrame(
  columns=['model', 'recall', 'mrr', 'ndcg', 'hit', 'precision', 'time']
)
```

```python
for model in model_list[0:20]:
  try:
    start = time.time()
    if model in model_neg_list:
      config_dict = {'neg_sampling': None, 'show_progress': False,}
    elif model in model_ract_list:
      config_dict = {'train_stage': 'actor_pretrain', 'pretrain_epochs': 150, 'neg_sampling': None, 'show_progress': False,}
    elif model in model_s3rec_list:
      config_dict = {'train_stage': 'pretrain', 'save_step': 10, 'show_progress': False,}
    else:
      config_dict = {'show_progress': False,}
    result = run_recbole(model=model, dataset=dataset, config_dict=config_dict, saved=False)
    t = time.time() - start
  except:
    continue
  test_result = result['test_result']
  recbole_result = recbole_result.append({
    'model': model, 
    'recall': test_result['recall@10'], 
    'mrr': test_result['mrr@10'], 
    'ndcg': test_result['ndcg@10'], 
    'hit': test_result['hit@10'], 
    'precision': test_result['precision@10'], 
    'time': round(t)
  }, ignore_index=True)
```

```python
from recbole.quick_start import run_recbole

dataset='ml-100k'
context_model_list = [
  # Context-aware recommendation 18
  'LR', 'FM', 'NFM', 'DeepFM', 'xDeepFM',
  'AFM', 'FFM', 'FwFM', 'FNN', 'PNN',
  'DSSM', 'WideDeep', 'DIN', 'DCN', 'AutoInt',
  'xgboost', 'lightgbm', 'DIEN',
]
```

```python
import time
import pandas as pd

context_result = pd.DataFrame(
  columns=['model', 'auc', 'logloss']
)
```

```python
for model in context_model_list[0:9]:
  try:
    start = time.time()
    config_dict = {'show_progress': False,}
    result = run_recbole(model=model, dataset=dataset, config_dict=config_dict, saved=False)
    t = time.time() - start
  except:
    continue
  test_result = result['test_result']
  context_result = context_result.append({
    'model': model, 
    'auc': test_result['auc'], 
    'logloss': test_result['logloss'], 
    'time': round(t)
  }, ignore_index=True)
```
